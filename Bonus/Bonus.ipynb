```python
%matplotlib inline
import numpy as np
from scipy import ndimage
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open('lena.jpg')
img
```

```python
print("img_mode: {}".format(img.mode),"\n" 
      "img_size: {}".format(img.size))
```

```python
img_gray = img.convert('L')
img_gray
```

```python
print("img_gray_mode: {}".format(img_gray.mode))
```

**將格式存為"*np.float64*"防止 data overflow**

```python
img_array = np.array(img_gray).astype(np.float64)
```

```python
plt.imshow(img_array, cmap='gray');
```

**Sobel Filter**

```python
sobelx = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]], dtype = np.float)
sobely = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]], dtype = np.float)
```

```python
#x axis
outx = ndimage.convolve(img_array, sobelx)
plt.imshow(outx, cmap='gray');
```

```python
#y axis
outy = ndimage.convolve(img_array, sobely)
plt.imshow(outy, cmap='gray');
```

```python
# both x and y
out = np.sqrt(outx*outx+outy*outy)
plt.imshow(out, cmap='gray');
```
